```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

playgrounds= pd.read_csv("https://data.wprdc.org/datastore/dump/47350364-44a8-4d15-b6e0-5f79ddff9367")
appearances= playgrounds.groupby('neighborhood').size()
countNeighborhoods= appearances.sort_values(ascending=False)
print(countNeighborhoods)
countNeighborhoods.plot()
plt.show()
```
