```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

playgrounds= pd.read_csv("https://data.wprdc.org/datastore/dump/47350364-44a8-4d15-b6e0-5f79ddff9367")
appearances= playgrounds.groupby('neighborhood').size()
countPlaygrounds= appearances.sort_values(ascending=False)
countPlaygrounds=countPlaygrounds.head(20)
plt.xticks(rotation=90)
countPlaygrounds.plot(kind="bar",x="neighborhood",y="count")
plt.tight_layout()
plt.show()
```

Shown by the data set, the "Best neighborhood" for playgrounds is Squirrel Hill South. That neighborhood has 8 total neighborhoods in it, which is three greater than the nearest neihgborhood, which is Beechview and South Side Slopes. Squirrel Hill South did turn out to be the "Best Neighborhood" for recreational activities.
