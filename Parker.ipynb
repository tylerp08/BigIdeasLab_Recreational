**Courts and Rinks**

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

courtsAndRinks = pd.read_csv("https://data.wprdc.org/datastore/dump/a5b71bfa-840c-4c86-8f43-07a9ae854227")
occurrences = courtsAndRinks.groupby('neighborhood').size()
sortedNeighborhoods = occurrences.sort_values(ascending=False)
print(sortedNeighborhoods)
courtsAndRinks['neighborhood'].value_counts().sort_index().plot(kind='bar', rot=90)
plt.rcParams['figure.figsize'] = [40, 15]
plt.show()
```

**Conclusion:**

When looking at the amount of courts and/or rinks in each neighborhood, it is clear that Squirrel Hill South has the most of any neighborhood, with 26 in total.  The two runner-ups of most courts and rinks are Highland Park with 20 and Hazelwood with 10, but when looking at the data or the graph, Squirrel Hill South towers above the other neighborhoods by a lot.  When taking into account the other datasets, it is determined that Squirrel Hill South is the best overall as well.
