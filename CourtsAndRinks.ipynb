**Courts and Rinks**

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

courtsAndRinks = pd.read_csv("https://data.wprdc.org/datastore/dump/a5b71bfa-840c-4c86-8f43-07a9ae854227")
occurrences = courtsAndRinks.groupby('neighborhood').size()
sortedNeighborhoods = occurrences.sort_values(ascending=False)
print(sortedNeighborhoods)
courtsAndRinks['neighborhood'].hist(bins=50)
plt.show()
```
