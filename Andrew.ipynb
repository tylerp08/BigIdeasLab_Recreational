```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

plt.figure(figsize=(12,4))
file_path = 'pools.csv'
pools = pd.read_csv(file_path)
poolsCount = pools.groupby('neighborhood').size()
sortedNieghborhoods = poolsCount.sort_values(ascending=False)
sortedNieghborhoods.plot()
plt.show()
```
